**PEP with Categorical Attributes**

```python
attri1 = pd.crosstab(df['gender'], df["pep"])
plt.show(attri1.plot(kind="bar"))
attri2 = pd.crosstab(df['region'], df["pep"])
plt.show(attri2.plot(kind="bar"))
```

From the gender by PEP barchart, it shows that the number of male-customer with PEP is greater than females but mostly they do nt prefer buying PEP. Moving to the barchart of region by PEP, it is seen that customers in the inner city seems to be the most active PEP buyers than people in town, rural and suburban respectively.

```python
attri3 = pd.crosstab(df['married'], df["pep"])
plt.show(attri3.plot(kind="bar"))
attri4 = pd.crosstab(df['car'], df["pep"])
plt.show(attri4.plot(kind="bar"))
```

For the barchart of married by PEP, it shows that married-customers have slightly higher number of buying PEP than a group of not married-customers. When looks at the car by PEP barchart, half of customers with PEP own cars and the rest of them don't.

```python
attri5 = pd.crosstab(df['savings_acct'], df["pep"])
plt.show(attri5.plot(kind="bar"))
attri6 = pd.crosstab(df['current_acct'], df["pep"])
plt.show(attri6.plot(kind="bar"))
attri7 = pd.crosstab(df['mortgage'], df["pep"])
plt.show(attri7.plot(kind="bar"))
```

With reference to the first barchart of savings_acct by PEP, it suggests that customers with savings account tend to be the one who buy PEP more than who do not have savings account. In addition, it also shows the similar trend in the current_acct by PEP barchart, people with current account buy PEP more than people who do not have. In contrast, customers with mortgage tend to buy PEP less than a group of customer without mortgage.

#### **3) Use z-score normalization to standardize the values of the income attribute. [Do not change the original income attribute in the table.]**


```python
income_z = (df["income"] - df["income"].mean()) / df["income"].std()
income_z.head(5)
```

```python
df = pd.concat([df,income_z.rename("income_z")], axis=1)
df.head()
```

#### **4) Discretize the age attribute into 3 categories (corresponding to "young", "mid-age", and "old"). [Do not change the original age attribute in the table.]**


```python
age_bins = pd.qcut(df.age, 3,labels=["young", "mid-age", "old"])
age_bins.head(5)
```

```python
df = pd.concat([df,age_bins.rename('age_bins')], axis=1)
df.head(5)
```

#### **5) Use Min-Max Normalization to transform the values of all numeric attributes  (income, age, children) in the original table (before the transforamtions in parts 3 and 4 above) onto the range 0.0-1.0.**


```python
age_norm = (df["age"] - df["age"].min())/(df["age"].max()-df["age"].min())
income_norm = (df["income"] - df["income"].min())/(df["income"].max()-df["income"].min())
children_norm = (df["children"] - df["children"].min())/(df["children"].max()-df["children"].min())

df = pd.concat([df,age_norm.rename('age_norm'),income_norm.rename("income_norm"),children_norm.rename('children_norm')], axis=1)
df.head()
```

#### **6) Convert the table (after normalization in part 5) into the standard spreadsheet format. Note that this requires converting each categorical attribute into multiple binary ("dummy") attributes (one for each values of the categorical attribute) and assigning binary values corresponding to the presence or not presence of the attribute value in the original record). The numeric attributes should remain unchanged. Save this new table into a file called bank_numeric.csv and submit it along with your assignment. [Hint: you might consider using the get_dummies for Pandas data frames.]**


```python
dummies = pd.get_dummies(df[["gender","region","married","car","savings_acct","current_acct","mortgage","pep"]])
dummies.head()
```

```python
df = pd.concat((df,dummies),axis=1)
df.head()
```

```python
df.to_csv('bank_numeric.csv')
```
